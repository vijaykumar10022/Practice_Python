```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris=load_iris()
# print(iris)
x=iris.data[:,2:]#petal length and widths
y=iris.target

tree_clf=DecisionTreeClassifier(max_depth=2)
tree_clf.fit(x,y)
```

```python
import pandas as pd
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/vijaykumar10022/Practice_Python/master/bank_authentication.csv")
data.head()
```

```python
x=data.iloc[:,0:4].values
y=data.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=3)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
cl=RandomForestClassifier(n_estimators=10,random_state=0)
cl.fit(X_train,Y_train)
y_pred=cl.predict(X_test)
```
